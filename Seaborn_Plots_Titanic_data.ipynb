Here, I work through a series of excercises supplied by Pierian Data and the Titanic data set from Seaborn to demonstrate the power of the grahical interface

## The Data

Like many of the Python-based moduls, Seaborn can import sample data sets. Here I demonstrate some code that generates visualization of the data with seaborn:


```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head() #Data columns
```

# Graphs

### Various Seaborn plots of passengers status and fate in the classic titanic data set 



```python
#Here I'm plotting age of the people that paid  fares to board the titanic

g= sns.jointplot(x = 'fare', y = 'age', data = titanic)
g.plot_joint(plt.scatter, color=".5", edgecolor="white")
g.plot_marginals(sns.distplot, kde=False, color=".5")
```

```python
# same plot as above in another color style
```

```python
#here I'm plotting the distribution of fares on the titanic. Pay attention to the number of bins used as this
#will impact your graph
sns.distplot(titanic['fare'], kde = False, bins = 40, color = 'black')
```

```python
#Here I'm showing a Boxplot of the class of ticket bought vs. age on the titanic
sns.boxplot(x="class", y="age", palette=["m", "g"],
            data=titanic)
```

```python
#example of a swarmplot
sns.swarmplot(x="class", y="age", data=titanic)
```

```python
sns.countplot(x = "sex", data = titanic)
```

### Below I show a heatmap of various passenger parameters and their survival

```python
sns.heatmap(titanic.corr(), cmap = 'coolwarm')
```

```python
#If you wanted to survive the titanic disaster, being female would have been an advantage
```

```python

g= sns.FacetGrid(data = titanic, col="sex", hue = "survived")
g.map(plt.hist,'age')
```

```python
g= sns.FacetGrid(data = titanic, col="sex")
g.map(plt.hist,'age')
```

```python
# You'd have been much more fortunate to have a first-class ticket on the titanic
g = sns.jointplot(data = titanic, x="pclass", y="survived", kind="kde", height=7, space=0)
```
