```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplot inline
```

```python
df = pd.read_csv('2019_nCoV_data.csv')
```

```python
df.head(30)
```

```python
df.describe()
```

```python
ax = sns.scatterplot(x='Province/State', y = 'Confirmed', data = df  )
ax.set_xticklabels(ax.get_xticklabels(),rotation=75)
```

```python
ax = sns.pairplot(df, palette="husl")
ax.set_xticklabels(ax.get_xticklabels(),rotation=75)
```

```python
ax = sns.violinplot(x='Date', y='Confirmed', data=df, pallette = 'RdBu')
ax.set_xticklabels(ax.get_xticklabels(),rotation=75
                   )
```
